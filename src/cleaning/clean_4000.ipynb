```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/raw/mobile_de_results_all.csv")
```

```python
df["price"] = (
    df["title"]
    .str.extract(r"(\d{1,3}(?:\.\d{3})*)\s*€")[0]
    .str.replace(".", "", regex=False)
    .astype(float)
)
```

```python
df = df.dropna(subset=["price"])
df["price"] = df["price"].astype(int)
```

```python
df.tail(10)
```

```python
df2 = df.copy()
```

```python
df2.drop('blocked',axis=1, inplace=True)
df2.drop('skipped',axis=1, inplace=True)
df2.drop('skip_reason',axis=1, inplace=True)
df2.drop('first_registration',axis=1, inplace=True)
df2.drop('price_eur',axis=1, inplace=True)
```

```python
df2=df2[['brand','model','price','km','year','title','url']]
```

```python
df2 = df2[df2["brand"].isin(df2["brand"].value_counts().loc[: "Jaguar"].index)]
```

```python
df2 = df2[df2["km"] < 155000]
```

```python
df2 = df2.dropna(subset=["price"])
```

```python
df2["price"] = df2["price"].astype(int)
```

```python
pd.set_option("display.max_rows", None)

df2["model"].value_counts()
```

```python
df_es = pd.read_csv("../../data/interim/spain_sample.csv", header=0)
```

```python
df_de = df2.copy()
df_de["country"] = "DE"
df_es["country"] = "ES"

df_all = pd.concat([df_de, df_es], ignore_index=True)
```
