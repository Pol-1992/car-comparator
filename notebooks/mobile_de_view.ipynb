```python
import pandas as pd

path = "../data/raw/mobile_de_FRESH.csv"
df = pd.read_csv(path, encoding="latin-1")
```

```python
df = df.reset_index().rename(columns={"index": "url"})
```

```python
df = df.drop(columns=["level_1","year","first_registration","skipped","skip_reason"])
```

```python
df = df.rename(columns={
    "blocked":"location",
    "location":"year",
    "fuel_type":"first_registration",
    "cv":"fuel_type",
    "km":"cv",
    "price_eur":"kW",
    "model":"km",
    "brand":"price",
    "title":"model",
    "url":"brand",
    "level_0":"url"
})
```

```python
df = df[[
    "brand",
    "model",
    "price",
    "km",
    "cv",
    "kW",
    "year",
    "first_registration",
    "fuel_type",
    "location",
    "url",
]]
```

```python
df["brand"] = df["brand"].str.replace("Patrocinado", "",regex=False)
```

```python
df["brand"] = df["brand"].str.replace("NUEVO", "",regex=False)
```

```python
df["brand"] = df["brand"].str.replace("CitroÃ«n", "Citroen", regex=False)
```
